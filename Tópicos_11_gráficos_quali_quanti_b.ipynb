Usaremos:

- **Quati (grupos)**: `sex` e `race`



````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Carregar o dataset Adult (Census Income) como DataFrame
adulto = fetch_openml('adult', version=2, as_frame=True).frame
````


##  **hours-per-week** por grupos

### Barras com **média** (sem barras de erro para simplificar)


````python
plt.figure(figsize=(6,4))
sns.barplot(data=adulto, x='sex', y='hours-per-week', estimator='mean', errorbar=None)
plt.title('hours-per-week (média) por sex')
plt.xlabel('sex'); plt.ylabel('hours-per-week')
plt.tight_layout(); plt.show()
````

````python
plt.figure(figsize=(7,4))
sns.barplot(data=adulto, x='race', y='hours-per-week', estimator='mean', errorbar=None)
plt.title('hours-per-week (média) por race')
plt.xlabel('race'); plt.ylabel('hours-per-week')
plt.tight_layout(); plt.show()
````
